## Probability distributions 概率分布。
### continuous random variables

```python
#scipy.stats library
# rvs: Random Variates
# pdf/pmf: Probability Density/Mass Function
# cdf: Cumulative Distribution Function
# stats: Return mean, variance, (Fisher’s) skew, or (Fisher’s) kurtosis
# moment: non-central moments of the distribution
import numpy as np
import matplotlib.pyplot as pl
%matplotlib inline
from scipy import stats as st
```

```python
mean, sd = 0,1
x = np.linspace(mean-6*sd, mean+6*sd, 100) #np.linspace主要用来创建等差数列
# start:返回样本数据开始点
# stop:返回样本数据结束点
# num:生成的样本数据量，默认为50
# endpoint：True则包含stop；False则不包含stop
# retstep：If True, return (samples, step), where step is the spacing between samples.(即如果为True则结果会给出数据间隔)
# dtype：输出数组类型
# axis：0(默认)或-1
x
```

```python
x.shape
```

```python
f = st.norm.pdf(x=x,loc = mean, scale = sd)
f
#scipy.stats.norm函数 可以实现正态分布（也就是高斯分布）
#norm.pdf(x, loc, scale) 等同于
# norm.pdf(y) / scale ，其中 y = (x - loc) / scale
# loc: mean 均值， scale: standard deviation 标准差
# 求概率密度函数指定点的函数值 
# stats.norm.pdf正态分布概率密度函数

# 求累计分布函数指定点的函数值 
# stats.norm.cdf正态分布累计概率密度函数

# 累计分布函数的逆函数
# stats.norm.ppf正态分布的累计分布函数的逆函数，即下分位点
```

```python
F = st.norm.cdf(x=x, loc=mean, scale=sd)
F
```

```python
pl.figure(figsize=(10,5))
pl.subplot(121)
pl.stem(x,f);pl.xlabel('x');pl.ylabel('f(x)');pl.title('pdf with mean = %.f and s.d. =%.f'%(mean,sd))
#pl.title('pdf with mean = {}').format(mean)) ;  {} ==%.?  ''.format() == ''%?
pl.subplot(122)
pl.stem(x, F);pl.xlabel('x');pl.ylabel('F(x)');pl.title('cdf with mean= {} and s.d. = {}'.format(mean,sd))
pl.show()
```

```python
mean, sd = 10, 0.025
x = np.linspace(mean - 6*sd, mean + 6*sd, 100)
f = st.norm.pdf(x=x, loc=mean, scale=sd)
F = st.norm.cdf(x=x, loc=mean, scale=sd)

pl.figure(figsize=(10,5))
pl.subplot(121)
pl.stem(x, f); pl.xlabel('x'); pl.ylabel('f(x)'); pl.title('pdf with mean={} and s.d.={}'.format(mean, sd))
pl.subplot(122)
pl.stem(x, F); pl.xlabel('x'); pl.ylabel('F(x)'); pl.title('cdf with mean={} and s.d.={}'.format(mean, sd))
pl.show()
```
