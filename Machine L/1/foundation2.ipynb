```python
import numpy as np
import matplotlib.pyplot as pl
import numpy as np
#从0-501   t = np.arange(501)
t = np.arange(501)/50
print(t[:4])
f = 1
y = np.sin(2*np.pi*f*t)*np.exp(-t/2)  #np相当于math+for循环
print(y[:4])

pl.scatter(t,y)  #横纵
pl.show()
```

```python
from matplotlib.widgets import Slider

ax = pl.subplot(111)
#print(ax) #AxesSubplot(0.125,0.11;0.775x0.77)
pl.subplots_adjust(left=0.25, bottom= 0.25)  #画布大小
t = np.arange(0.0,10,0.02)
a0 = 0
f0 = 1
s = np.exp(a0*t)*np.sin(2*np.pi*f0*t)
l,  = pl.plot(t,s, lw = 2, color = 'red') #蓝色波浪线 ,lw 加粗，color = ‘’
axfreq = pl.axes([0.25, 0.1, 0.65, 0.03])  #第三个是长，四个是高,前两个是位置。
axdamp  = pl.axes([0.25, 0.15, 0.65, 0.03])

sfreq = Slider(axfreq,'Freq', 0.5, 3.0, valinit=f0)#valinit填充的长度
sdamp = Slider(axdamp, 'Damp', -0.5, 0.5, valinit=a0)



def update(val):
    damp = sdamp.val
    freq = sfreq.val
    l.set_ydata(np.exp(damp*t)*np.sin(2*np.pi*freq*t))
    pl.draw()
sfreq.on_changed(update)
sdamp.on_changed(update)  #感受不到变化

pl.show()
```
